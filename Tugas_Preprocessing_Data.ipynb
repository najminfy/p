```python
import pandas as pd
import numpy as np

df = pd.read_csv('movie_sample_dataset.csv')
df['color']=df['color'].str.title()
df.head()
```

```python
#menghilangkan tanda negatif
df['duration']=df['duration'].abs()
df['imdb_score']=df['imdb_score'].abs()
#cek banyaknya missing value pada masing-masing kolom
df.replace(["","N/A","Null"], np.nan, inplace = True)
print(df.isnull().sum())
```

```python
#replace missing data color dengan modus
mode_color=df['color'].mode()[0]
print("Mode of Color:",mode_color)
df['color'].replace(np.nan, mode_color, inplace=True)
```

```python
#replace missing data director name dengan modus
mode_director=df['director_name'].mode()[0]
print("Mode of director:",mode_director)
df['director_name'].replace(np.nan, mode_director, inplace=True)
```

```python
#drop missing data gross by row
df.dropna(subset=["gross"], axis=0, inplace=True)
```

```python
print(df[["genres","director_name","movie_title"]][df["genres"].isnull()])
print(df["genres"][df['director_name']=='Christopher Spencer'].value_counts())
print(df["genres"][df['movie_title']=='Son of God'].value_counts())
```

karena tidak ada nama direktor dan judul film serupa maka tidak jadi replace data genres by frequency

```python
#replace missing data genres dengan modus
mode_genres=df['genres'].mode()[0]
print("Mode of genres:",mode_genres)
df['genres'].replace(np.nan, mode_genres, inplace=True)
```

```python
#drop missing data budget by row
df.dropna(subset=["budget"], axis=0, inplace=True)
```
