```python
import pandas as pd
import numpy as np

#membaca file data training dan menampilkan informasi umum
data_train=pd.read_csv('data_training.csv')
print(data_train.info())

#membaca file data testing dan menampilkan informasi umum
data_test=pd.read_csv('data_testing.csv')
print(data_test.info())
```

```python
data_train
```

```python
#membagi data menjadi variabel x dan y
x=data_train.drop(['quality','Id'],axis=1)
y=data_train['quality']
```

```python
#menormalisasikan data menjadi rentang 0-1 agar seimbang saat pelatihan model
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
#membagi data training menjadi data train dan test untuk membuat dan melatih model prediksi
#20% data digunakan untuk membuat model prediksi dan 80% digunakan untuk menguji seberapa akurat modelnya
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#membuat model prediksi dari x dan y data train dengan metode decision tree
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
#memprediksi nilai x test
#digunakan untuk membandingkan dengan y test
y_pred=classifier.predict(x_test)
```

```python
#evaluasi model prediksi
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
data_test
```

```python
#mendefinisikan variabel x pada file data testing
x_testing=data_test.drop('Id',axis=1)
x_testing=pd.DataFrame(scaler.fit_transform(x_testing),columns=x_testing.columns)
x_testing
```

```python
#menggunakan model prediksi pada file data testing
y_predict=classifier.predict(x_testing)
```

```python
#membuat dataset baru yang berisi ID dan hasil prediksi Quality
df=pd.DataFrame({'Id': data_test['Id'], 'Quality': y_predict})
df.head()
```
